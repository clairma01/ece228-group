```python
import os
import sys
import glob
import numpy as np
```

```python
data_path = "/content/drive/MyDrive/Deepfish/"
```

```python
def check_text_file(file_path):
  return(file_path.endswith(".txt"))
```

```python
train_outfile = open(data_path+"train.txt", "a")
test_outfile = open(data_path+"valid.txt", "a")

for directory in os.listdir(data_path):
  if not(check_text_file(directory)):
    for sub_directory in os.listdir(os.path.join(data_path, directory)):
      if not(check_text_file(sub_directory)):
        for image in os.listdir(os.path.join(data_path, directory, sub_directory)):
          if not(check_text_file(image)):
            if (sub_directory == "train"):
              train_outfile.write(os.path.join(data_path, directory, sub_directory, image))
              train_outfile.write("\n")
            elif (sub_directory == "valid"):
              test_outfile.write(os.path.join(data_path, directory, sub_directory, image))
              test_outfile.write("\n")
```

```python
#adding no fish images to file
neg_data_path = "/content/drive/MyDrive/Deepfish/Negative_samples/"
neg_samples = []

for folder in os.listdir(neg_data_path):
  neg_dir_samples = []
  for image in os.listdir(os.path.join(neg_data_path, folder)):
    if not(check_text_file(image)):
      neg_dir_samples.append(os.path.join(neg_data_path, folder, image))
  neg_samples.append(neg_dir_samples)
```

```python
train_split = 0.7

for i in range(len(neg_samples)):
  train_samples = int(np.ceil(train_split * len(neg_samples[i])))
  test_samples = int(np.floor((1-train_split)*len(neg_samples[i])))
  for j in range(train_samples):
    train_outfile.write(neg_samples[i][j])
    train_outfile.write("\n")
  for k in range(test_samples):
    test_outfile.write(neg_samples[i][k + train_samples])
    test_outfile.write("\n")

train_outfile.close()
test_outfile.close()
```
